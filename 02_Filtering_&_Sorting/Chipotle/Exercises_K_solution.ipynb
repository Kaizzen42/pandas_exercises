### Step 3. Assign it to a variable called chipo.

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
chipo = pd.read_csv(url, sep='\t')
chipo
```

### Step 4. How many products cost more than $10.00?

```python
chipo['item_price_float'] = chipo['item_price'].str.replace('$','').astype('float')
chipo['item_price_float_per_item'] = chipo['item_price_float'] / chipo['quantity']
item_groups = chipo.groupby('item_name')
max_cost = item_groups.max()
max_cost[max_cost['item_price_float_per_item']>10.00].shape[0]
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
chipo[['item_name','item_price_float_per_item']]
```

### Step 6. Sort by the name of the item

```python
chipo.sort_values(by=['item_name'])
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
most_expensive_item = max_cost[max_cost['item_price_float_per_item'] == max(max_cost['item_price_float_per_item'])]
most_expensive_item
# chipo['quantity'][]
```
